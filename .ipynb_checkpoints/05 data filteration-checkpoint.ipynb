## LOC AND ILOC

```python
d = {
    'name': ['karunesh', 'kanu', 'bansal'],
    'age': [19,20,25]
}
```

```python
df = pd.DataFrame(d, index=['A', 'B', 'C'])
```

```python
df
```

## FILTERING BY INDEX NAME

```python
df.loc['A'] # when we select row it use square bracket in this we game name of index
```

```python
df.loc[['A', 'B']]
```
